```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fx(x):
    return 3 * x ** 2 - 3 * x + 4

def fpx(x):
    return 6 * x - 3


# coding challenge

def fx0(x):
    return np.cos(2 * np.pi * x) + x ** 2

def fpx0(x):
    return -2 * np.pi * np.sin(2 * np.pi * x) + 2 * x
```

```python
x = np.linspace(-3, 3, 1000)

plt.plot(x, fx(x), label='f(x)')


plt.plot(x, fpx(x), label="f'(x)")
plt.xlabel('x')
plt.ylabel('f(x)')

plt.grid()
plt.legend()
plt.show()
```

```python
def gradient_descent(fx, fpx, x0, learning_rate=0.1, max_iter=1000):
    x = x0
    x_values = [x]
    driver_values = [fpx(x)]
    for i in range(max_iter):
        x_new = x - learning_rate * fpx(x)
        
        x_values.append(x_new)
        driver_values.append(fpx(x_new))
        x = x_new
        
    return x_new, np.array(x_values), np.array(driver_values)
```

```python
x0 = np.random.choice(np.linspace(-3, 3, 1000))

x_min, x_values, driver_values = gradient_descent(fx, fpx, x0, learning_rate=0.01, max_iter=100, tol=1e-6)
print(x_min)
```

```python
plt.plot(x, fx(x), label='f(x)')

plt.plot(x, fpx(x), label="f'(x)")


plt.plot(x_min, fx(x_min), 'ro')
plt.plot(x_min, fpx(x_min), 'ro')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
x0 = np.random.choice(np.linspace(-3, 3, 1000))

x_min, x_values, driver_values = gradient_descent(fx, fpx, x0, learning_rate=0.001, max_iter=100000000, tol=1e-6)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x_values,'o-')
ax[0].set_xlabel('Iteration')
ax[0].set_ylabel('x')
ax[0].set_title(f"x_min = {x_min}")
ax[0].grid()


ax[1].plot(driver_values, 'o-')
ax[1].set_xlabel('Iteration')
ax[1].set_ylabel("f'(x)")
ax[1].set_title(f"f'(x_min) = {fpx(x_min)}")
ax[1].grid()

plt.show()
```
